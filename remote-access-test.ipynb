_Import libraries to open MySQL file and manipulate databases_

```python
import pandas as pd
import pymysql
```

```python
#Modify parameterse accordingly

host="XXXXX"
port=3306
dbname="i519TestDB"
user="axmcc"
password="XXXXXX"

conn = pymysql.connect(host, user=user,port=port,
                           passwd=password, db=dbname)
```

_Import graphing library_

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(18,6))
pandas_df=pd.read_sql('SELECT hire_date FROM EMPLOYEES;', con=conn) 
pandas_df.groupby(pandas_df.hire_date).size().plot(kind='bar')
plt.xlabel('Hire Date')
plt.ylabel('# of observations')
plt.show()
```
